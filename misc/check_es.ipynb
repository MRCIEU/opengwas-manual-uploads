````python
from elasticsearch import Elasticsearch
````

````python
gwas_id='GCST90016997'
index='ebi-a'

es = Elasticsearch(
	[{'host': 'localhost','port': '9200'}],
)

body={
	"query":{
		"term":{"gwas_id":gwas_id},
	}
}
es.count(index=index, body=body)
````

````python
gwas_id='GCST90016997'
index='ebi-a-4'

es = Elasticsearch(
	[{'host': 'localhost','port': '9200'}],
)

body={
	"query":{
		"term":{"snp_id":"rs234"},
	}
}
a = es.search(index=index, body=body, size=10000)
a
len(a['hits']['hits'])
````

````python
# a['hits']['hits']
[x['_source']['gwas_id'] for x in a['hits']['hits']]
````

````python

body={
	"query":{
		"bool":{
			"filter":
				[
					{"term":{"gwas_id":gwas_id}},
					{"term":{"snp_id":"rs234"}}
				]
		}
	}
}
es.count(index=index, body=body)
es.search(index=index, body=body)
````
